```python
# py imports
import os
import sys
import glob
from argparse import ArgumentParser
import time
import matplotlib.pyplot as plt

# third party
import tensorflow as tf
import scipy.io as sio
import numpy as np
from keras.backend.tensorflow_backend import set_session
from scipy.interpolate import interpn
import scipy.ndimage

import keras.backend as K
from keras.models import Model
import keras.layers as KL
from keras.layers import Layer
from keras.layers import Conv3D, Activation, Input, UpSampling3D, concatenate, MaxPooling3D, AveragePooling3D
from keras.layers import LeakyReLU, Reshape, Lambda, PReLU, add, ReLU, Softmax, multiply, subtract
from keras.initializers import RandomNormal
import keras.initializers

# project
sys.path.append('./ext/')
import medipy
import networks
from medipy.metrics import dice
import datagenerators
import neuron.layers as nrn_layers
import neuron.models as nrn_models
import neuron.utils as nrn_utils


def nn_trf_all(vol_size, interp_method='nearest'):

    ndims = len(vol_size)
    
    # Inputs
    subj_input = Input(shape=[*vol_size, 1])
    flow_1 = Input(shape=[*vol_size, ndims])
    
    Down_size = [int(i*1/2) for i in vol_size]
    flow_2 = Input(shape=[*Down_size, ndims])
    
    Down_size = [int(i*1/4) for i in vol_size]
    flow_3 = Input(shape=[*Down_size, ndims])
    
    Down_size = [int(i*1/8) for i in vol_size]
    flow_4 = Input(shape=[*Down_size, ndims])
    
    Down_size = [int(i*1/16) for i in vol_size]
    flow_5 = Input(shape=[*Down_size, ndims])

    # transform
    warp_vol_1 = nrn_layers.SpatialTransformer(interp_method=interp_method, indexing='ij')([subj_input, flow_1])  
    
    flow = networks.trf_resize(flow_2, 1/2)
    warp_vol_2 = nrn_layers.SpatialTransformer(interp_method=interp_method, indexing='ij')([subj_input, flow]) 
    
    flow = networks.trf_resize(flow_3, 1/4)
    warp_vol_3 = nrn_layers.SpatialTransformer(interp_method=interp_method, indexing='ij')([subj_input, flow]) 
    
    flow = networks.trf_resize(flow_4, 1/8)
    warp_vol_4 = nrn_layers.SpatialTransformer(interp_method=interp_method, indexing='ij')([subj_input, flow]) 
    
    flow = networks.trf_resize(flow_5, 1/16)
    warp_vol_5 = nrn_layers.SpatialTransformer(interp_method=interp_method, indexing='ij')([subj_input, flow]) 
    
    return Model(inputs=[subj_input, flow_1, flow_2, flow_3, flow_4, flow_5], outputs=[warp_vol_1, warp_vol_2, warp_vol_3, warp_vol_4, warp_vol_5])
```

```python
test_dir = '../../Data/Brain-MRI-npz/Testing/Buckner40/'
label = np.load(test_dir+'label.npy')
test_pair = np.load(test_dir+'test_pairs.npy')[64]
location_1 = 80
location_2 = 110
print(test_pair)


fixed = bytes.decode(test_pair[0])
moving = bytes.decode(test_pair[1])
fixed_vol, fixed_seg = datagenerators.load_example_by_name(test_dir+fixed)
moving_vol, moving_seg = datagenerators.load_example_by_name(test_dir+moving)
fixed_vol = fixed_vol[0,...,0]
moving_vol = moving_vol[0,...,0]


plt.imshow(fixed_vol[:,:,location_1],cmap='Greys_r',vmax=1)
plt.show()
plt.imshow(fixed_vol[:,location_2,:],cmap='Greys_r',vmax=1)
plt.show()

plt.imshow(moving_vol[:,:,location_1],cmap='Greys_r',vmax=1)
plt.show()
plt.imshow(moving_vol[:,location_2,:],cmap='Greys_r',vmax=1)
plt.show()
```

```python
load_model_file = 'models/DeepCumulative_L5/90.h5'

# Prepare
vol_size = [144,192,160]       


# device handling
device = '/gpu:0'
config = tf.ConfigProto()
config.gpu_options.allow_growth = True
config.allow_soft_placement = True
set_session(tf.Session(config=config))


fixed = bytes.decode(test_pair[0])
moving = bytes.decode(test_pair[1])
fixed_vol, fixed_seg = datagenerators.load_example_by_name(test_dir+fixed)
moving_vol, moving_seg = datagenerators.load_example_by_name(test_dir+moving)

mov_down_2 = scipy.ndimage.interpolation.zoom(moving_vol, (1,0.5,0.5,0.5,1), order = 1)
mov_down_4 = scipy.ndimage.interpolation.zoom(moving_vol, (1,0.25,0.25,0.25,1), order = 1)
mov_down_8 = scipy.ndimage.interpolation.zoom(moving_vol, (1,0.125,0.125,0.125,1), order = 1)
mov_down_16 = scipy.ndimage.interpolation.zoom(moving_vol, (1,0.0625,0.0625,0.0625,1), order = 1)


# load weights of model
with tf.device(device):
    model = networks.DeepCumulative_L5(vol_size)
    model.load_weights(load_model_file)

    # NN transfer model
    nn_trf_model_nearest = nn_trf_all(vol_size, interp_method='nearest')
    nn_trf_model_linear = nn_trf_all(vol_size, interp_method='linear')


with tf.device(device):
            
    pred = model.predict([fixed_vol, moving_vol, mov_down_2, mov_down_4, mov_down_8, mov_down_16])
    
    warp_vol_1, warp_vol_2, warp_vol_3, warp_vol_4, warp_vol_5 = nn_trf_model_linear.predict([moving_vol, pred[0], pred[2], pred[4], pred[6], pred[8]])
    warp_seg_1, warp_seg_2, warp_seg_3, warp_seg_4, warp_seg_5 = nn_trf_model_nearest.predict([moving_seg, pred[0], pred[2], pred[4], pred[6], pred[8]]) 

    
Dice_vals = dice(warp_seg_1[0,...,0], fixed_seg[0,...,0], label)              
print('Dice mean: {:.3f} ({:.3f})'.format(np.mean(Dice_vals), np.std(Dice_vals)))

Dice_vals = dice(warp_seg_2[0,...,0], fixed_seg[0,...,0], label)              
print('Dice mean: {:.3f} ({:.3f})'.format(np.mean(Dice_vals), np.std(Dice_vals)))

Dice_vals = dice(warp_seg_3[0,...,0], fixed_seg[0,...,0], label)              
print('Dice mean: {:.3f} ({:.3f})'.format(np.mean(Dice_vals), np.std(Dice_vals)))

Dice_vals = dice(warp_seg_4[0,...,0], fixed_seg[0,...,0], label)              
print('Dice mean: {:.3f} ({:.3f})'.format(np.mean(Dice_vals), np.std(Dice_vals)))

Dice_vals = dice(warp_seg_5[0,...,0], fixed_seg[0,...,0], label)              
print('Dice mean: {:.3f} ({:.3f})'.format(np.mean(Dice_vals), np.std(Dice_vals)))

Dice_vals = dice(moving_seg[0,...,0], fixed_seg[0,...,0], label)              
print('Dice mean: {:.3f} ({:.3f})'.format(np.mean(Dice_vals), np.std(Dice_vals)))
```
